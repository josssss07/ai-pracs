# Linear regression

```python
#1: get all the needed imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# 2. Create a simple dataset (house sizes and house prices)
# House sizes (in square feet)
x = np.array([1500, 1600, 1700, 1800, 1800, 2000, 2100, 2200, 2300, 2400]).reshape(-1, 1)

# Corresponding house prices (in thousands of dollars)
y = np.array([300, 320, 340, 360, 380, 400, 420, 440, 460, 480])
```

```python
#3: split the data into training and test data
x_train, x_test , y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#4 create and train the linear regression model 
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#5: predict using the given model
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Prediction is: {y_pred}')
print(f"mean squared error: {mse}")
```

```python
#print the test values vs the predicted values
print("Test Values vs Predicted Values:")
for actual, predicted in zip(y_test, y_pred):
    print(f"Actual: {actual}, Predicted: {predicted:.2f}")
```
