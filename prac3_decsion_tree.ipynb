```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib as plt
```

```python
#read the csv data
df = pd.read_csv('/home/josh/Dev/tyit_sem5/pracs_final/ai/csv/data1.csv')

#assign weights to the data frame
d = {'UK': 0, 'USA': 1, 'N': 2}

#take a deciding factor
df['Nationality'] = df['Nationality'].map(d)

#add the yes/ no weights
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

#select the features
features = ['Age', 'Experience', 'Rank', 'Nationality']

#make the x and y plot
X = df[features]
y = df['Go']

#run the library for the dtree
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

#plot the dtree
tree.plot_tree(dtree, feature_names=features)
```
