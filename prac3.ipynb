```python
#step 1: import all the libraries 
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
#step 2: load the data and make the test and train data
iris = load_iris()
df = pd.DataFrame(data=iris, columns= iris.feature_names)
df['target'] = iris.target
```

```python
features = iris.feature_names
x = df[features]
y = df['target']

#split into train and test
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42) 
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
y_pred = dtree.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'accuracy is: {accuracy * 100:.2f}') 
```

```python
tree.plot_tree(dtree, feature_names=features)
```

```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Step 2: Load the Iris dataset
iris = load_iris()
X = iris.data    # Features
y = iris.target  # Labels (target)

# Step 3: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Step 5: Plot the decision tree
plt.figure(figsize=(12, 8))  # Setting figure size
tree.plot_tree(clf, 
               feature_names=iris.feature_names, 
               class_names=iris.target_names, 
               filled=True,  # Colors the nodes according to their class
               rounded=True) # Rounds the corners of the nodes
plt.show()
```
